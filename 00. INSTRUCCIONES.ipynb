# Ejecución de los Notebooks.

El proyecto está compuesto por un total de 6 Notebook enumerados del 01 al 06. Cada uno de estos archivos lleva en su nombre el índice correspondiente Ej: 01. Nombre_notebook.ipynb. Los Notebooks deben ejecutarse en el orden asignado por la numeración. A continuación se enlistan los archivos mensionados:

* 01. Cargue y exploración de datos
* 02. Gestión de datos desbalanceados y selección de estrategia
* 03. Evaluación de hiperparámetros y selección del mejor modelo
* 04. Preprocesamiento del dataset para ampliar la información de entrenamiento
* 05. Ejecución del mejor modelo con el dataset ampliado
* 06. Reducción de dimensionalidad y prueba de desempeño

Los notebook se encuentran en el repositorio: https://github.com/cristhiant24/EACD-HomeCreditDefaultRisk


# Acceso y lectura de datos

La información para la ejecución del proyecto fue suministrada por “Home Credit” por medio de una convocatoria de competencia de predicción realizada en Kaggle.com en el año 2018. El acceso a los mismos está disponible en la siguiente URL: https://www.kaggle.com/competitions/home-credit-default-risk/data


Adicional a los datos suministrados por “Home Credit” desde Kaggle, se construyó un archivo llamado “type_variables” en el cual se hace una descripción de cada una de las variables y este archivo se usa en el preprocesamiento de los datos. Este archivo se encuentra en el repositorio https://github.com/cristhiant24/EACD-HomeCreditDefaultRisk




El código está elaborado con una estructura en la cual lee y exporta los datos a una única carpeta. Todos los ficheros mencionados deben estar en una misma carpeta y a un mismo nivel. En los notebooks esta un objeto llamado “ruta_lectura”, en él se encuentra la ruta de esta carpeta, la cual se debe ajustar en caso de querer replicar la ejecución del código desde otra espacio de guardado.  En el proceso de ejecución de los distintos notebooks se van generando y guardando distintos ficheros de datos que van rumbo a la carpeta relacionada en “ruta_lectura”.
